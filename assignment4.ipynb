# DATA DISTRIBUTIONS(ASSIGNMENT4)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal_dist = np.random.normal(60, 17, 100) 
```

```python
plt.hist(normal_dist, bins = 10, color = 'gold');
plt.axvline(normal_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(normal_dist.mean() + normal_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(normal_dist.mean() - normal_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**This normal distribution variable as we can see on the plot,the mean is at the center of the symetric sides 
and we can also see that the standard deviation is almost equally appart on both sides.Therefore we can have 
a visual understanding that discribitive statistics can usefully explain our data distribution.**

```python
chi_dist = np.random.chisquare(3, size= 100)
```

```python
plt.hist(chi_dist, color = 'gold');
plt.axvline(chi_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(chi_dist.mean() + chi_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(chi_dist.mean() - chi_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**In chi distribution with small degrees of freedom, like in the above case and as we can see from the histodram,
the distribution can not be usefully explained by the mean and standard deviation because of its skewness.But with larger degrees of freedom we might.**

```python
exp_dist = np.random.exponential(scale= 1.0, size= 100)
plt.hist(exp_dist, color = 'gold');
plt.axvline(exp_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(exp_dist.mean() + exp_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(exp_dist.mean() - exp_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**We have a skewed distribution, mean and standard deviation can not usefully explain the exponential distribution.**

**lets have a look at another distribution;

```python
logistic_dist = np.random.logistic(loc=0.0, scale= 1.0, size = 100)
plt.hist(logistic_dist, color = 'gold');
plt.axvline(logistic_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(logistic_dist.mean() + logistic_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(logistic_dist.mean() - logistic_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**As we can see from our histogram in the above,the distribution is aproximately symetrical;therefore mean and 
standard deviation can usefully explain the distribution.**

```python
geometric_dist = np.random.geometric(.42, size = 100)
plt.hist(geometric_dist, color = 'gold');
plt.axvline(geometric_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(geometric_dist.mean() + geometric_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(geometric_dist.mean() - geometric_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**mean and standard deviation can not explain the distribution.**

```python
uniform_dist = np.random.uniform(low =0.0, high =1.0, size= 100)
plt.hist(uniform_dist, color = 'gold');
plt.axvline(uniform_dist.mean(), color='m', linestyle='solid', linewidth=2,marker = 's');
plt.axvline(uniform_dist.mean() + uniform_dist.std(), color='m', linestyle='--', linewidth=2, marker = 's');
plt.axvline(uniform_dist.mean() - uniform_dist.std(), color='m', linestyle='--', linewidth=2,marker = 's');
```

**mean and standard deviation can usefully explain the distribution.**
