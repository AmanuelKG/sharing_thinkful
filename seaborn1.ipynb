**1.**<br/>**A. Four different plots using 'mpg' column from mpg dataset**
- Frequency distribution of 'mpg' variable.Its pros is it gives us the overview of the quantitative variables distribution shape about normal little bit skewed to left.Considering its main purpose it has done its job.

```python
sns.set(style = 'darkgrid')
sns.distplot(mpg['mpg'], color = 'g', bins = 20,kde = False,norm_hist = False)
plt.ylabel('Frequency',fontsize = 15)
plt.xlabel('MPGs',fontsize = 15)
plt.title('The MPG Frequency Distribution', fontsize = 15, fontweight = 'bold')
sns.despine(left = True, bottom = True)
```

- Probability density plot using kernel density estimation(KDE) plot and rug plot.Its pros is it gives us the probability density function plot for variable 'mpg' and the shape of the curve.and the rug plot collaborates the shape of the kde plot and us we can see most of the data points are between 20-35.Even if we have a histogram plot the kde plot is more clearer in view to check the normality of a data.Using all 3 helps to understand a variable more solidly.

```python
sns.set(style = 'darkgrid')
sns.distplot(mpg['mpg'], color = 'g', hist = False,rug = True)
plt.ylabel('Density',fontsize = 15)
plt.xlabel('MPGs',fontsize = 15)
plt.title('The MPG Distribution using kde and rug plots', fontsize = 15, fontweight = 'bold')
sns.despine(left = True, bottom = True)
```

- using Boxplot to explore the mpg variable helps to visualize the dispersion of the data.From the boxplot below we can observe the median(center) and the 25% and 75% of the data and the outlier.The cons of this plot is that its limitation to tel the normality of the distribution.But it can do a great job in collaboration with the distribution plots.

```python
sns.set(style = 'darkgrid')
sns.boxplot(y = 'mpg', data = mpg,color = 'g', width = 0.3)
plt.ylabel('MPGs',fontsize = 15)
plt.title('mpg Boxplot',fontsize = 15, fontweight = 'bold');
```

- The pros of this plot is,so that we can have an everview of all the datapoints and their concentration through out the values of the variable 'mpg' at a glance.Its cons are,we can have very limited statistical information and if the sample size is high data overlaps can make it unclear to understand the real dispersion of the data. 

```python
sns.set(style = 'darkgrid')
sns.catplot(y = 'mpg', data = mpg, color = 'g',kind = 'strip');
```

**B.Three different plots using continuous variables 'displacement' and	'horsepower',columns of data from mpg dataset.**

```python
g = sns.lmplot(y='displacement', # Variable 1.
               x='horsepower', # Variable 2.
                   data=mpg, # Data
               fit_reg= True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.5})
g.set_ylabels("Displacement",fontweight = 'bold')
g.set_xlabels("Horsepower",fontweight = 'bold')
plt.title('Scatter plot featuring displacement and horsepower',fontsize = 15,fontweight = 'bold');
```

pros of the above plot:
- The scatter plot shows how the two variables are related
- the linear regration model line helps to understand if the relationship is linear.<br/>
cons:
  - it fails to address the distribution of the indivisual quantitative variables

```python
g = sns.jointplot(y='displacement',x ='horsepower',data = mpg, kind = 'reg')
g.ax_joint.set_ylabel("Displacement",fontweight = 'bold')
g.ax_joint.set_xlabel("Horsepower",fontweight = 'bold');
plt.title('Scatter plot featuring displacement and horsepower',fontsize = 15,fontweight = 'bold', pad = 70);
```

pros of the above plot:
- it shows the distribution of each quantitative variable using histogram
- The scatter plot shows how the two variables are related
- the linear regration model line helps to understand if the relationship is linear.

```python
g = sns.jointplot(y='displacement',x ='horsepower',data = mpg, kind = 'kde')
g.ax_joint.set_ylabel("Displacement",fontweight = 'bold')
g.ax_joint.set_xlabel("Horsepower",fontweight = 'bold');
plt.title('Scatter plot featuring displacement and horsepower',fontsize = 15,fontweight = 'bold', pad = 70
          ,loc='right');
```

Pros of the above plot:
- it shows the overall strength of relationship concentration around certain data points than others.
- it can help visualization to understand the spread of datapoints<br/>
cons:
- it narrows the width  of the spread and difficult to understand if there is far dispersed datapoints relationships.

**C.I Chose one continuous variable 'acceleration' and one categorical variable 'cylinders', and plotting them six different ways.**

```python
sns.stripplot(x = 'cylinders',y = 'acceleration', data = mpg,jitter=False);
```

the data points in the above plot are overlapped.therefore to have a clearer view of the data points lets use the 
default value of jitter which is true.

```python
sns.stripplot(x = 'cylinders',y = 'acceleration', data = mpg);
```

swarmplots gives much clearer data point display than the jittered stripplot.

```python
sns.swarmplot(x = 'cylinders',y = 'acceleration', data = mpg);
```

```python
sns.boxplot(x = 'cylinders',y = 'acceleration', data = mpg);
```

```python
sns.violinplot(x = 'cylinders',y = 'acceleration', data = mpg);
```

**Comparing the two above Plots:**<br/>
The first plot is a boxlpot chich shows the the three quartiles,the whiskers and outliers.and in the second plot
we have a box plot on top of the violinplot. Unlikely the box plot, the violin plot allows us to actually plot 
all the components that corresponds to actual data points. The violin plot is essentially showing the kernel density 
estimation of the underlying distribution.

```python
sns.barplot(x = 'cylinders',y = 'acceleration', data = mpg);
plt.title('Mean Acceleration for each cylinders',fontsize = 12,fontweight = 'bold');
```
