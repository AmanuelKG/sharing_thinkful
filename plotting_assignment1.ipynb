**First thing first i imported all the packages that i will need for the process**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

**The next step is loading the data to this jupyter notebook for purposes of initial visualization and understanding the structure of the data.**

```python
Data = pd.read_csv('purchases.csv')
```

```python
#lets see the overview of the data
Data.info()
```

```python
# This dataframe has 4 columns and 10 rows of data,so lets have a look at how the data is structured in tables.
Data.head()
```

```python
#To help us understand more the structure and the pattern of the data let me plot one of the 
# quantitative columns in a simple line plot.I have included the data points using dot marker.
#the line plot give a general adea as how the data points are changing with this data index.
plt.plot(Data['ad_views'], color = 'magenta', marker = 'o', alpha = 0.5);
plt.ylabel('ad_Views');
plt.xlabel('The Data Index Value');
plt.title("Viewed advertisments");
```

**Now lets see what both the quantitative columns i.e "ad_views" and "items_purchased"**
**look like plotted side by side, i will plot them both in the same canvas for a close look.**

```python
plt.plot(Data['ad_views'], color = 'red', marker = '^');
plt.plot(Data['items_purchased'], color = 'green', marker = '*');
plt.legend();
```

*From the plot you can have a general understanding about the spread of data in both variables.*
*Just by visualizing the plot 'ad_views' are more spreaded in value than 'items_purchsed'.*

**Lets explore our data more by using different visualization plotting tools.Lets use Histogram now.Histogram is 
a frequecy plot.I am plotting both plots in one figure as subplots side by side**

```python
plt.figure(figsize = (10,5));
plt.subplot(1,2,1);
plt.hist(Data['ad_views'], color = 'yellow', bins = 15);
plt.ylabel('Frequency');
plt.xlabel('Add Views');  

plt.subplot(1,2,2);
plt.hist(Data['items_purchased'], color = '#0000ff');
plt.xlabel('Items Purchased');
plt.tight_layout();
```

**From the left yellow histogram we can see that there is a very high frequency(almost 3 times) of**
**add Views between 10 to 20, around 15.But there is equal frequency of only 1 in the other data** 
**points range.This is because we have only 10 sample elements in our data,which is too few to be a sample data set.**
**we can read almost similiar frequency level in the second histogram as well for almost the same reason except for** 
**the frequency at around five is 2.**

**Lets move on and let me continue plotting a catogorical one this time.Bar will be one of the players**
**in this business;i will plot a bar plot of 'ad_views' agains 'country'.**

```python
plt.bar(data = Data,x = 'country', height = 'ad_views', alpha = 0.5);
plt.ylabel('Add Views');
plt.xlabel('Country');
```

**From the bar plot we can visualize that there are more add viewers in the US than in Canada.And by setting .**
**alpha to 0.5 we can see all the data points in each countries and their spread in the number of views per person.**


**So far we have gone through a number of different plots,and now it is time to try the scatter plot,which is good 
**for pai plotting to understand the overview of a relationship between variables(columns' data).Let me illustrate 
**its practical functionality using its plot.*** 

```python
plt.scatter(data = Data, x = 'ad_views', y = 'items_purchased', marker = '*', s = 50);
plt.ylabel('Items Purchased');
plt.xlabel('add Views');
```

**Because of the size of the data it is quite difficulty to read the relationship between the two columns.
**This is just for illustration purposes of plotting scatter plot.but if the size of our data was suffiecient
**we could have seen some patterns of relationship not neccessarly though.**

**Lastly but not list i will finish my todays assignment by plotting box plot to read general distribution of the**
**data around its quartiles.Using this plot we may have a general understanding of the outliers we expect from 
**our data.**

```python
plt.figure(figsize = (10, 5));
plt.subplot(1, 2, 1);
plt.boxplot(x = Data['ad_views'], patch_artist = True);
plt.ylabel('Add Views');

plt.subplot(1,2,2);
plt.boxplot(x = Data['items_purchased'], patch_artist = True);
plt.ylabel('Items Purchased');
```

**From the plots we can see at a glance that the median of added views is between 20 and 30, and 
**the median items purchased is between 2 and 3. And also we can read that 25th percentile of added views is between 
**10 - 20(the lower side of the box) and 75th percentile is between 50-60 (the upper side of the box).We can also 
**see that there are no outliers in the data because there are no dots beyond the upper and lower whiskers.
**Similiar concept can be infered from the items purchased plot.***
