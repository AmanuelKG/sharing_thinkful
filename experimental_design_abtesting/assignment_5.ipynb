### A/B Testing and T-tests

1. $\bar{y1}$ =5 ,  $\bar{y2}$ =8 ,  ğ‘ 1=1 ,  ğ‘ 2=3 ,  ğ‘1=200 ,  ğ‘2=500

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
sample_1 = np.random.normal(5, 1, 200)
```

```python
sample_2 = np.random.normal(8, 1, 500)
```

```python
stats.ttest_ind(sample_1, sample_2)
```

2. $\bar{y1}$ =1090 , $\bar{y2}$ =999, ğ‘ 1=400, ğ‘ 2=30, ğ‘1=900, ğ‘2=100
